### Statistical Functions

```python
def chi_square_test(df, alpha):
    rows = len(df.axes[0])
    cols = len(df.axes[1])
    expected_df = pd.DataFrame(index=range(rows),columns=range(cols))
    test = 0
    for i in range(rows):
        for j in range(cols):
            row_sum = df.iloc[i].sum()
            col_sum = df.iloc[:, j].sum()
            total_sum = df.values.sum()
            expected_df[j][i] = ((row_sum * col_sum) / total_sum)
            diff = df.loc[i][j] - expected_df.loc[i][j]
            sq_diff = diff*diff
            test += (sq_diff/expected_df.loc[i][j])
    deg_free = (rows-1) * (cols-1)
    
    print("Alpha: ", alpha, " Degrees of Freedom: ", deg_free, " Test: ", test)
```

Finally, it is shown that if a protest had violence, then it was unlikely to be accomodated. This might help give support to the idea that peaceful protests are more likely to be effective in getting a good governmental response.

```python
def prob_accuracy(guesses, labels):
    total = 0
    correct = 0
    for i in range(len(guesses)):
        n = labels['response_amount'].loc[i].astype(int)
        y = knn.predict_proba(X_test)[i]
        y = np.argpartition(y, -n)[-n:]
        y = [i + 1 for i in y]
        y.sort()
        
        x = [labels['killings'][i],
             labels['accomodation'][i]*2,
             labels['shootings'][i]*3,
             labels['crowd dispersal'][i]*4,
             labels['arrests'][i]*5,
             labels['ignore'][i]*6,
             labels['beatings'][i]*7]

        x = [i for i in x if i != 0]
        
        
        if x == y:
            correct += 1
            
        total += 1
        
        
    return (correct/total) * 100
    
```

##### Binary Classification for Each Potential Response

```python
cols_to_drop = ['killings', 'accomodation', 'shootings', 'crowd dispersal', 'arrests', 'ignore', 
                'beatings', 'response_amount', 'response']

classes = ['killings', 'accomodation', 'shootings', 'crowd dispersal', 'arrests', 'ignore', 'beatings']

accuracies = {}
predictions = []
for response in classes:
    y = knn_protest[response]
    X = knn_protest.drop(cols_to_drop, axis = 1)

    SEED = 45
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=SEED)
    
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    max_acc = 0
    best_n = 0
    best_knn = KNeighborsClassifier(n_neighbors=0)
    
    for i in range(1,10):
        n = i
        knn = KNeighborsClassifier(n_neighbors=n)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        
        if accuracy > max_acc:
            max_acc = accuracy
            best_n = n
            best_knn = knn
        
    
    accuracies[response] = ((max_acc)*100, best_n)
    predictions.append(y_pred)

predictions_updated = []
for i in range(len(predictions[0])):
    predictions_updated.append([predictions[0][i], predictions[1][i], predictions[2][i], predictions[3][i], 
                             predictions[4][i], predictions[5][i], predictions[6][i]])
    


y = knn_protest[classes]
X = knn_protest.drop(cols_to_drop, axis = 1)

SEED = 45
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=SEED)


y_test = y_test.reset_index()

real_values = []
for i in range(y_test.shape[0]):
    real_values.append([y_test['killings'].loc[i], 
                        y_test['accomodation'].loc[i],
                        y_test['shootings'].loc[i],
                        y_test['crowd dispersal'].loc[i],
                        y_test['arrests'].loc[i],
                        y_test['ignore'].loc[i],
                        y_test['beatings'].loc[i]])
    
correct = 0
total = 0
for i in range(len(real_values)):
    if real_values[i] == predictions_updated[i]:
        correct += 1
    total += 1

accuracy = (correct/total) * 100

print("Accuracy: ", accuracy)
```
